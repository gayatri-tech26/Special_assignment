```python
sns.histplot(df['age'])         #shows the count of the patients.
```

```python
df['sex'].value_counts().plot(kind='bar')
```

```python
plt.scatter(df['age'], df['bmi'])
plt.xlabel('age')
plt.ylabel('bmi')
plt.show()
```

```python
plt.hexbin(df['age'],df['charges'],gridsize=10)
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
region_count = df['region'].value_counts()
plt.bar(region_count.index, region_count.values)
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_sorted = df.sort_values('age')

plt.plot(df_sorted['age'], df_sorted['charges'])
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
region_count = df['region'].value_counts()
plt.pie(region_count.values,
        labels=region_count.index,
        autopct='%1.1f%%')
plt.title('Customer Distribution by Region')
plt.show()
```
